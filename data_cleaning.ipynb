```python
pss_cols = ['Unable_to_control','Stressed', 'Handling_problems', 'Going_right', 'Not_coping','Controlling_irritations', 'On_top', 'Angered','Difficulties_in_overcoming']
```

```python
# Reverse positively worded items
reverse_items = ['Unable_to_control','Stressed', 'Handling_problems', 'Going_right', 'Not_coping','Controlling_irritations', 'On_top', 'Angered','Difficulties_in_overcoming']
for col in reverse_items:
    social_media2[col + '_rev'] = 4 - social_media2[col]

# List all items for descriptive stats
pss_items_for_analysis = ['Unable_to_control','Stressed', 'Handling_problems', 'Going_right', 'Not_coping','Controlling_irritations', 'On_top', 'Angered','Difficulties_in_overcoming']
```

```python
# Descriptive statistics
pss_desc = social_media2[pss_items_for_analysis].describe().T  # transpose for readability
pss_desc = pss_desc[['count', 'mean', 'std', 'min', '25%', '50%', '75%', 'max']]
pss_desc.rename(columns={'std':'SD', '50%':'Median'}, inplace=True)

print(pss_desc)
```

```python
# Calculate mean and SD
pss_stats = social_media2[pss_cols].agg(['mean','std']).T
pss_stats.columns = ['Mean','SD']

# Identify highest and lowest-rated items
highest_item = pss_stats['Mean'].idxmax()
lowest_item = pss_stats['Mean'].idxmin()

print(pss_stats)
print(f"\nHighest-rated stress symptom: {highest_item} (Mean = {pss_stats.loc[highest_item,'Mean']:.2f})")
print(f"Lowest-rated stress symptom: {lowest_item} (Mean = {pss_stats.loc[lowest_item,'Mean']:.2f})")
```
